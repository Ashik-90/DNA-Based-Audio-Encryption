```python
x,sr = librosa.load(file_path)
plt.title('dog.wav')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.figure(figsize=(10,5))
librosa.display.waveshow(x,sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(10,5))
librosa.display.specshow(Xdb, sr=sr)
plt.colorbar()
plt.title('Decibel-scaled spectrogram of encrypted file')
```

```python
from librosa import load

# Load two audio signals
f1, sr = load('abcd.wav')
f2, sr = load('decrypted.wav')

# Compute cross-correlation using numpy.correlate
corr = np.correlate(f1, f2, mode='same')

#norm_corr = corr / (np.std(f1) * np.std(f2))

# Plot the cross-correlation
plt.title('Correlation: Original and Decrypted')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.plot(corr)
plt.show()
```
